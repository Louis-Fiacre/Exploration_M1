# I. a. Importe l'ensemble des librairies nécéssaires

```python
import os
import pandas as pd
from collections import Counter
```

```python
cwd = '/home/lf/Bureau/Memoire/2_Exploitation/out_spacy'
cwd_out = '/home/lf/Bureau/Memoire/2_Exploitation/Visualisation'

if os.path.isdir(cwd):
    
    files_list = sorted(os.listdir(cwd))
    dfs = list()
    
    for file_name in files_list:
        
        file_path = os.path.join(cwd,file_name)
        if file_path.endswith('.xlsx'):
            df = pd.read_excel(file_path, index_col=[0,1])
            dfs.append(df)
            print(file_name)
    
```

#  Similarité des documents


```python
import matplotlib.pyplot as plt
%matplotlib inline
import sklearn
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
#instanciation
sc = StandardScaler()
```

```python
itera = 0
for df in dfs:
    
    name =(files_list[itera]).split('.xlsx')[0]
    
    X = df
    n, p = X.shape[0], X.shape[1]
    Z = sc.fit_transform(X)
    acp = PCA(svd_solver='full')
    coord = acp.fit_transform(Z)

    #positionnement des individus dans le premier plan
    fig, axes = plt.subplots(figsize=(12,12))
    axes.set_xlim(-10,10) #même limites en abscisse
    axes.set_ylim(-10,10) #et en ordonnée
    #placement des étiquettes des observations
    legend = list()

    for i in range(Z.shape[0]):
        x,y = coord[i,0],coord[i,1]
        # color par pamphlet ou non
        auteur = df.index[i][1][0:2]

        if auteur in ['GB','LB', 'JB', 'VH']:
            color_a = 'red'
        else:
            color_a = 'blue'

        plt.plot(x,y,marker="o", markersize=10, color=color_a)
        plt.annotate(i,[x,y])
        legend.append(str(i)+' '+df.index[i][1])

    #ajouter les axes et legende
    lege = axes.legend(legend, bbox_to_anchor=(1,0.75))
    axes.set_title(name,fontsize=15)
    plt.plot([-10,10],[0,0],color='silver',linestyle='-',linewidth=1)
    plt.plot([0,0],[-10,10],color='silver',linestyle='-',linewidth=1)

    #affichage et save
    plt.savefig(os.path.join(cwd_out,'ACP_'+name+'.png'), format='png', facecolor='white',bbox_extra_artists=(lege,), bbox_inches='tight')
    plt.show()
        

    itera+=1
```

### Dendrogramme
## Classification Ascendante Hiérarchique, avec l'indice de Ward 

```python
from scipy.spatial.distance import squareform, pdist 
from scipy.cluster.hierarchy import linkage, dendrogram
```

```python
itera = 0
for df in dfs:
    
    X = df
    name =(files_list[itera]).split('.xlsx')[0]

    n, p = X.shape[0], X.shape[1]

    Z = sc.fit_transform(X)
    acp = PCA(svd_solver='full')
    coord = acp.fit_transform(Z)

    dist = squareform(pdist(coord))
    
    fig = plt.figure(figsize=(10, 5))
    
    dend = dendrogram(linkage(dist, method='ward'), leaf_rotation=90, labels=df.index.get_level_values(1))
    title_name = 'Dendrogramme_'+name
    plt.title(title_name)
    
    plt.savefig(os.path.join(cwd_out,title_name+'.png'),format='png', facecolor='white', bbox_inches='tight')
    plt.show()
    itera+=1
```
